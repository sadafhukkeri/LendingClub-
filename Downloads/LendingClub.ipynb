```python
# define a function to cap outliers in numeric columns
def cap_outliers(df, lower_quantile=0.01, upper_quantile=0.99):
    # select only numeric columns (float and int)
    df_num = df.select_dtypes(include=['float64', 'int64']).copy()

    # loop through each numeric column
    for col in df_num.columns:
        # get the lower and upper quantile values
        lower = df_num[col].quantile(lower_quantile)
        upper = df_num[col].quantile(upper_quantile)
        # clip values outside the quantile range
        df_num[col] = df_num[col].clip(lower, upper)

    # update the original dataframe with the capped values
    df.update(df_num)
    return df

# apply the outlier capping function to the dataframe
df = cap_outliers(df)

# get a list of all numeric column names
num_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# plot boxplots for numeric columns in groups of 5
for i in range(0, len(num_cols), 5):
    # select a batch of 5 columns
    batch = num_cols[i:i+5]
    # create boxplots for the batch
    df[batch].plot(kind='box', subplots=True, layout=(1, len(batch)),
                   figsize=(4 * len(batch), 5), sharey=False)
    # set a title for the plot
    plt.suptitle(f'Boxplots for columns {i + 1} to {i + len(batch)}', fontsize=14)
    # adjust layout to avoid overlap
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    # show the plot
    plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer

# get names of numeric columns from the dataframe
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# create the transformer using yeo-johnson method
pt = PowerTransformer(method='yeo-johnson')

# create an empty dictionary to store the transformed columns
transformed_data = {}

# go through each numeric column one by one
for col in numeric_cols:
    # drop missing values for fitting the transformer
    col_data = df[[col]].dropna()

    # skip the column if all values are the same
    if col_data[col].nunique() <= 1:
        print(f"Skipping '{col}' - constant values")
        continue

    try:
        # create a new transformer for the column
        pt_col = PowerTransformer(method='yeo-johnson')

        # apply the transformation
        transformed = pt_col.fit_transform(col_data)

        # create a new series to match the original dataframe's shape
        full_col = pd.Series(index=df.index, dtype='float64')

        # place the transformed values into the correct positions
        full_col[col_data.index] = transformed.flatten()

        # save the result to the dictionary
        transformed_data[col] = full_col

    except Exception as e:
        # skip the column if there’s an error and print the reason
        print(f"Skipping '{col}' - {e}")
```

**Balancing the dataset before training the model**

> Using oversampling SMOTE ( Synthetic Minority Over-sampling Technique ) synthetic samples are not duplicates of existing data but are new data points created between real data points using interpolation



```python
from imblearn.over_sampling import SMOTE
from collections import Counter

print('before SMOTE : ', Counter(y_train))

smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train,y_train)
print('after SMOTE : ',Counter(y_train))
```

```python
logistic_reg = LogisticRegression()
```

```python
# parameters which we defined are initialized in gridsearchcv we have passed here logistic model , parameters,scoring for performance matrics
# cv=5 is cross validation it will divide data into 5 parts and 4 parts will be used as training and 1 part as testing
# it will repeat for 5 times
grid_cv = GridSearchCV(logistic_reg, param_grid, scoring='accuracy',cv=5)
```

# XGBoost


> It's a type of boosting algorithm that builds multiple decision trees sequentially, where each tree try to fix errors of previous one. It focuses on improving performance step by step.



```python
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb

xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    use_label_encoder=False,
    n_jobs=-1,
    random_state=42
)

# defining the parameter grid to search
param_dist = {
    'booster': ['gbtree'],
    'learning_rate': [0.01, 0.03, 0.05, 0.1],
    'max_depth': [3, 4, 5, 6],
    'min_child_weight': [1, 2, 3],
    'gamma': [0, 1, 2, 3],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'n_estimators': [100, 150, 200, 300]
}

# random search object
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=30,  # number of random combinations to try
    scoring='roc_auc',
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# Fit to training data
random_search.fit(X_train, y_train)
random_search.best_params_
```

```python
random_search.best_score_
```

**SHAP (SHapley Additive exPlanations)**


> It creates an explainer for model, calculates SHAP values for the test set, and then plots summary to show top 5 most important features in the prediction



```python
import shap

# create a SHAP explainer for the XGBoost model
explainer = shap.TreeExplainer(best_model)

# calculate SHAP values for the test data, without checking additivity
shap_values = explainer(X_test, check_additivity=False)

# create a summary plot to visualize the most important features (max 5 features displayed)
shap.summary_plot(shap_values, X_test, max_display=5)
```

By observing the SHAP value :

Risk goes up when -
*   out_prncp_inv (outstanding principal invested) is high
*   recoveries and collection_recovery_fee are high

Risk goes down when -
*   total_rec_prncp (total principal received) is high
*   last_pymnt_amnt is high
